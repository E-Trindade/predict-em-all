### Insights

 - Obviously, most of the caught pokemons are in the common classification
 - **#122 (Mr Mime)** is the most common very rare pokemon, **#088 (Grimer)** is second.
 - **New York City** has the highest number of rare pokemons
 - Prague has the highest number of very rare pokemons even if most of the data came from America
 - Europe have more very rare pokemons compared to other continents
 - Most of the very rare pokemons appeared during **Saturdays and Fridays**.
This was also observed in common and rare pokemons.
Hence, this is just because the **players are playing more frequently during those days**.
This is also observable for other time related variables.
 - Rare and very rare pokemons are mostly found in terrainType = 13.
This is maybe because **terrainType=13** is the most common terrainType
 - Rare pokemons are **highly distributed**, very rare pokemons are **more concentrated**
 - Local Pokemons are apparent (#122 - Mr Mime - are only found in Europe, ... ); #130 (Gyarados) are concentrated; ...
 - Distribution of very rare pokemons are the same regardless if the location is closeToWater or not
 - Some very rare pokemons in **New York** are caught only when close to water. Some are caught only when not close to water. **#131 (Lapras)** and **#137 (Porygon)** are only caught when close to water. #113 (Chansey) is only caught when not close to water
 - Looking at the specific pokemon Lapras, it was not consistently caught only if close to water. It was only consistent in New York
 - Distribution of very rare pokemons are the same regardless of **weather**
 - In **Asia**, all **#142 (Aerodactyl)** were caught not close to water.
 - I can add more if needed.

```r
# Some Exploratory Data Analysis

# Obviously, most of the caught pokemons are in the common classification
ggplot(p.data, aes(x=classif)) +
  geom_bar() + 
  labs(x="classification") +
  ggtitle("Distribution Over Different Classifications")

# #122 (Mr Mime - exclusive pokemon in Europe) is the most common very rare pokemon, #088 (Grimmer) is second.
ggplot(p.very.rare, aes(x=class)) +
  geom_bar() +
  coord_flip() +
  ggtitle("Very Rare Pokemons Distribution Over Different Classes")

# New York Ciry has the highest number of rare pokemons
ggplot(p.rare, aes(x=city)) +
  geom_bar() +
  coord_flip() +
  ggtitle("Rare Pokemons Distribution Over Different Cities")

# Prague has the highest number of very rare pokemons even if most of the data came from America
ggplot(p.very.rare, aes(x=city)) +
  geom_bar() +
  coord_flip() +
  ggtitle("Very Rare Pokemons Distribution Over Different Cities")

# Europe have more very rare pokemons compared to other continents
ggplot(p.very.rare, aes(x=continent)) +
  geom_bar() +
  coord_flip() +
  ggtitle("Very Rare Pokemons Distribution Over Different Continents")

# Most of the very rare pokemons appeared during Saturdays and Fridays.
# This was also observed in common and rare pokemons
# Hence, this is just because the players are playing more frequently during those days.
# This is also observable for other time related variables
ggplot(p.very.rare, aes(x=appearedDayOfWeek)) +
  geom_bar() +
  coord_flip() +
  ggtitle("Very Rare Pokemons Distribution Over Different Days of a Week")

# Rare and very rare pokemons are mostly found in terrainType = 13.
# But this is maybe because terrainType=13 is the most common terrainType
ggplot(p.no.common, aes(x=as.factor(p.no.common$terrainType))) +
  geom_bar() +
  coord_flip() +
  ggtitle("Very Rare and Rare Pokemons Distribution Over Different Terrain Types")
  
# Rare pokemons are highly distributed, very rare pokemons are more concentrated
ggplot(p.no.common, aes(x=longitude, y=latitude, color=classif)) +
  geom_point() +
  ggtitle("Very Rare and Rare Pokeomons Distribution")

# Distribution of very rare pokemons are the same regardless if the location is closeToWater or not
ggplot(p.very.rare, aes(x=closeToWater, fill=class)) +
  geom_bar() +
  coord_flip() +
  scale_fill_brewer(palette="Spectral") +
  ggtitle("Very Rare Pokemons Distribution When Close to Water or Not")

# Some very rare pokemons in New York are caught only when close to water. Some are caught only when not close to water
# Pokemon 131 (Lapras) and 137 (Porygon) are only caught when close to water. Pokemon 113 (Chansey) is only caught when not close to water
ggplot(p.very.rare.NewYork, aes(x=closeToWater, fill=class)) +
  geom_bar() +
  coord_flip() + 
  ggtitle("Distribution of Very Rare Pokemons When Close to Water and When Not Close to Water in New York") +
  scale_fill_brewer(palette="Spectral")

# Looking at the specific pokemon Lapras, it was not consistently caught only if close to water. It was only consistent in New York
ggplot(p.very.rare.Lapras, aes(x=closeToWater)) +
  geom_bar() +
  coord_flip() + 
  ggtitle("Distribution of Lapras When Close to Water and When Not Close to Water") +
  scale_color_brewer(palette="Spectral")

# Distribution of very rare pokemons are the same regardless of weather
ggplot(p.very.rare, aes(x=weatherIcon, fill=class)) +
  geom_bar() +
  coord_flip() + 
  ggtitle("Distribution of Very Rare Pokemons on Different Weather Icons") +
  scale_fill_brewer(palette="Spectral")

# In Asia, all #142 (Aerodactyl) were caught not close to water.
ggplot(p.very.rare.Asia, aes(x=closeToWater, fill=class)) +
  geom_bar() +
  coord_flip() +
  ggtitle("Distribution of very rare pokemons In Asia When Close to Water and When Not Close to Water") +
  scale_fill_brewer(palette="Spectral")
```
